```python
transactions = pd.read_csv('./data/transactions_v2.csv')
transactions.head()
```

```python
print(transactions['msno'].nunique())
```

```python
print(transactions.shape)
transactions.info()
transactions.describe(include='all')

```

```python
col = ['transaction_date', 'membership_expire_date']
transactions[col] = transactions[col].apply(lambda x: pd.to_datetime(x, format='%Y%m%d'))
transactions.info()
```

```python
# Tomamos solo usuarios del train
train_users = set(train["msno"])

train_txn = transactions[transactions["msno"].isin(train_users)].copy()

# Para cada usuario, tomamos su última expiración conocida
user_cutoff = (
    train_txn
    .groupby("msno")["membership_expire_date"]
    .max()
    .rename("cutoff_date")
    .reset_index()
)

print(user_cutoff.shape)
print(user_cutoff.head())
print(user_cutoff["msno"].nunique())
```

```python
transactions["payment_method_id"] = transactions["payment_method_id"].astype("int16")
transactions["payment_plan_days"] = transactions["payment_plan_days"].astype("int16")
transactions["plan_list_price"] = transactions["plan_list_price"].astype("int32")
transactions["is_cancel"] = transactions["is_cancel"].astype("int8")
transactions["is_auto_renew"] = transactions["is_auto_renew"].astype("int8")
```

```python
txn_valid = transactions[transactions["is_cancel"] == 0]
```

```python
txn_count = transactions.groupby("msno").size().rename("txn_count")
txn_count.head()
print(txn_count.max())
```

```python
valid_txn_count = txn_valid.groupby("msno").size().rename("valid_txn_count")
```

```python
print(transactions['msno'].nunique())
```

```python
last_txn = txn_valid.groupby("msno")["transaction_date"].max()
last_txn.head()
```

```python
recency = (cutoff_date - last_txn).dt.days.rename("txn_recency_days")
recency.head()
```

```python
cancel_count = transactions.groupby("msno")["is_cancel"].sum().rename("cancel_count")
has_cancelled = (cancel_count > 0).astype(int).rename("has_cancelled")
print(cancel_count.head())
print(has_cancelled.head())
print(transactions['is_cancel'].value_counts())
```

```python
cancel_ratio = (cancel_count / txn_count).rename("cancel_ratio")
```

```python
print(cancel_count.value_counts())
```

```python
print(transactions['is_auto_renew'].value_counts())
```

```python
auto_renew_count = transactions.groupby("msno")["is_auto_renew"].sum().rename("auto_renew_count")
has_auto_renew = (auto_renew_count > 0).astype(int).rename("has_auto_renew")
print(auto_renew_count.head())
```

```python
print(auto_renew_count.value_counts())
```

```python
plan_days_median = txn_valid.groupby("msno")["payment_plan_days"].median().rename("plan_days_median")
print(plan_days_median.head())
print(plan_days_median.value_counts())
```

```python
last_plan_days = txn_valid.sort_values("transaction_date") \
    .groupby("msno")["payment_plan_days"] \
    .last().rename("last_plan_days")

print(last_plan_days.head())
print(last_plan_days.value_counts())
```

```python
avg_price = txn_valid.groupby("msno")["plan_list_price"].mean().rename("avg_plan_price")
print(avg_price.head())
print(avg_price.value_counts())
```

```python
txn_features = pd.concat([
    txn_count,
    valid_txn_count,
    recency,
    cancel_count,
    has_cancelled,
    auto_renew_count,
    has_auto_renew,
    plan_days_median,
    last_plan_days,
    avg_price,
], axis=1).reset_index()
```

```python
print(txn_features.head().to_string())
print(txn_features.shape)
```
