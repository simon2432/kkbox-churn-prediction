```python

df = transactions.copy()
t = pd.to_datetime(df['transaction_date'])
e = pd.to_datetime(df['membership_expire_date'])
delta_dias = (e - t).dt.days
plan = df['payment_plan_days']
valido = t.notna() & e.notna() & plan.notna()
cumple = valido & (delta_dias == plan)
no_cumple = valido & (delta_dias != plan)
sin_datos = ~valido
print(f"Filas con fechas y plan válidos: {valido.sum()}")
print(f"Cumplen (delta_dias == payment_plan_days): {cumple.sum()}")
print(f"No cumplen: {no_cumple.sum()}")
print(f"Sin comparar (NaN en fecha o plan): {sin_datos.sum()}")
```

```python
n_dup = transactions.duplicated().sum()
print(f"Filas duplicadas (segunda en adelante): {n_dup}")

print(f"Filas únicas: {transactions.drop_duplicates().shape[0]} de {len(transactions)}")

dupes = transactions[transactions.duplicated(keep=False)]
print(f"Filas en grupos con duplicados: {len(dupes)}")
print(dupes.head(20))  
```

```python
# 1. Quedarnos solo con transactions de usuarios presentes en base
transactions_base = transactions[transactions['msno'].isin(base['msno'])].copy()

print(f"Filas de transactions para usuarios de base: {len(transactions_base):,}")
print(f"Usuarios únicos en transactions_base: {transactions_base['msno'].nunique():,}")
print(f"Usuarios únicos en base: {base['msno'].nunique():,}")
```

```python
cutoff = '2017-02-28'

transactions_before_cutoff = transactions_base[
    transactions_base['transaction_date'] <= cutoff
]
```

```python
txn_features = (
    transactions_before_cutoff
    .groupby('msno')
    .agg(
        txn_count=('transaction_date', 'count'),
        txn_last_date=('transaction_date', 'max'),
        txn_first_date=('transaction_date', 'min'),
        cancel_count=('is_cancel', 'sum'),
        has_cancelled=('is_cancel', 'max'),
        auto_renew_last=('is_auto_renew', 'last'),
        payment_plan_days_last=('payment_plan_days', 'last'),
        plan_list_price_last=('plan_list_price', 'last'),
        actual_amount_paid_last=('actual_amount_paid', 'last'),
        payment_plan_days_median=('payment_plan_days', 'median'),
        actual_amount_paid_median=('actual_amount_paid', 'median'),
    )
    .reset_index()
)

txn_features['days_between_first_last_txn'] = (
    txn_features['txn_last_date'] - txn_features['txn_first_date']
).dt.days

txn_features.head()
```

```python
df_model = base.merge(txn_features, on='msno', how='left')
```

```python
df_model['has_txn_history'] = df_model['txn_count'].notnull().astype(int)
```

```python
df_model['txn_count'] = df_model['txn_count'].fillna(0)
df_model['cancel_count'] = df_model['cancel_count'].fillna(0)
df_model['has_cancelled'] = df_model['has_cancelled'].fillna(0)
```
